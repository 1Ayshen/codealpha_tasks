```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("books.csv")

df.head()
```

```python
rating_order = ["One", "Two", "Three", "Four", "Five"]

rating_counts = (
    df["Rating"]
    .value_counts()
    .reindex(rating_order)
)

plt.figure(figsize=(8,5))

plt.bar(rating_counts.index, rating_counts.values)

plt.title("Book Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Number of Books")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Price"], bins=10)

plt.title("Distribution of Book Prices")
plt.xlabel("Price (£)")
plt.ylabel("Number of Books")

plt.show()
```

```python
plt.figure(figsize=(6,5))

plt.boxplot(df["Price"])

plt.title("Book Price Box Plot")
plt.ylabel("Price (£)")

plt.show()
```

```python
average_price = (
    df.groupby("Rating")["Price"]
      .mean()
      .reindex(rating_order)
)

plt.figure(figsize=(8,5))

plt.bar(average_price.index, average_price.values)

plt.title("Average Book Price by Rating")
plt.xlabel("Rating")
plt.ylabel("Average Price (£)")

plt.show()
```

```python
plt.figure(figsize=(7,7))

plt.pie(
    rating_counts,
    labels=rating_counts.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Rating Distribution")

plt.show()
```

```python
top_books = df.nlargest(10, "Price")

plt.figure(figsize=(10,6))

plt.barh(top_books["Title"], top_books["Price"])

plt.title("Top 10 Most Expensive Books")
plt.xlabel("Price (£)")
plt.ylabel("Book Title")

plt.show()
```

```python
cheap_books = df.nsmallest(10, "Price")

plt.figure(figsize=(10,6))

plt.barh(cheap_books["Title"], cheap_books["Price"])

plt.title("Top 10 Cheapest Books")
plt.xlabel("Price (£)")
plt.ylabel("Book Title")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df["Page"], df["Price"])

plt.title("Book Price vs Page")
plt.xlabel("Page")
plt.ylabel("Price (£)")

plt.show()
```

```python
df["Cumulative Price"] = df["Price"].cumsum()

plt.figure(figsize=(10,5))

plt.plot(df["Cumulative Price"])

plt.title("Cumulative Book Prices")
plt.xlabel("Book Index")
plt.ylabel("Cumulative Price (£)")

plt.show()
```

```python
price_bins = [0, 20, 40, 60]
labels = ["Low", "Medium", "High"]

df["Price Category"] = pd.cut(
    df["Price"],
    bins=price_bins,
    labels=labels
)

df["Price Category"].value_counts().plot(kind="bar")

plt.title("Books by Price Category")
plt.xlabel("Category")
plt.ylabel("Number of Books")

plt.show()
```
